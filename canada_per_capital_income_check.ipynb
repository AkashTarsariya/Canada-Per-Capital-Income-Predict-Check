```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
X = df[['year']]
y = df['per capita income (US$)']
X = X.values.reshape(-1, 1)

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

## Train Regression Model

### Subtask:
Build and train a linear regression model using the prepared 'year' and 'per capita income (US$)' data.


```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, y)

print("Linear Regression model trained successfully.")
```

## Predict Per Capital Income for 2020

### Subtask:
Use the trained regression model to predict the per capita income for the year 2020.


```python
year_2020 = np.array([[2020]])
income_2020 = model.predict(year_2020)

print(f"Predicted per capital income for 2020: ${income_2020[0]:,.2f}")
```

## Visualize Regression and Prediction

### Subtask:
Create a scatter plot of the original data points and overlay the regression line. Clearly mark the predicted per capita income for 2020 on the plot. Ensure all axes and the legend are clearly labeled.


```python
plt.figure(figsize=(10, 6))

# 1. Create a scatter plot...
plt.scatter(X, y, color='blue', label='Original Data')

# 2. Plot the regression line
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# 3. Mark the predicted per capital income for 2020
plt.scatter(year_2020, income_2020, color='green', marker='X', s=150, label=f'Predicted 2020 Income: ${income_2020[0]:,.2f}')

# 4. Set the x-axis label as 'Year' and the y-axis label as 'Per Capita Income (US$)'
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')

# 5. Add a title to the plot...
plt.title('Canada Per Capital Income Prediction')

# 6. Include a legend...
plt.legend()

# 7. Display the plot...
plt.grid(True)
plt.show()
```
